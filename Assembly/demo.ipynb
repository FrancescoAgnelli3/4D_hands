```python
import pickle as pkl
import os
import numpy as np

path = "/mnt/wwn-0x5000000000000001-part1/datasets/AssemblyHands/assembly101-download-scripts/data/HAND_GCN/RAW_contex30_thresh0/train"
path_tgt = "/mnt/wwn-0x5000000000000001-part1/datasets/AssemblyHands/assembly101-download-scripts/data_our"

for file in os.listdir(path):
    if file.endswith(".pkl"):
        with open(os.path.join(path, file), 'rb') as f:
            data = pkl.load(f)
            data = np.array(data)
        length = file.split("_")[4]
        rgb = file.split("_")[5]
        file = file.replace(length, "_")
        file = file.replace(rgb, "_")
        file = file.replace("_____", "_")
        # save the data as numpy file in path_tgt
        np.save(os.path.join(path_tgt, file.replace(".pkl", ".npy")), data)
        
```
