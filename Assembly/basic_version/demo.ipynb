```python
import torch
import numpy as np

path = "/mnt/wwn-0x5000000000000001-part1/datasets/AssemblyHands/assembly101-download-scripts/data_processed/test_loader.pt"

data = torch.load(path)
task = []
for el in data:
    task.append(el[5])

# Example vector with n values between 0 and 11
vector = np.array(task)

# Count occurrences of each label
unique, counts = np.unique(vector, return_counts=True)

# Calculate percentages
percentages = {label: (count / len(vector)) * 100 for label, count in zip(unique, counts)}

# Display percentages
for label, percentage in percentages.items():
    print(f"Label {label}: {percentage:.2f}%")

print("Length of the dataset:", len(vector))
```
